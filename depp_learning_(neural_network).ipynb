```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/Data_sets/breast cancer.csv')
df
```

```python
df.isna().sum()
```

```python
df.drop(columns='Unnamed: 32',inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['diagnosis']=le.fit_transform(df['diagnosis'])
df
```

```python
X=df.iloc[:,2:]
X
```

```python
y=df['diagnosis']
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
minmax=MinMaxScaler()
minmax.fit(X_train)
X_train=minmax.transform(X_train)
X_test=minmax.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
ann=tf.keras.models.Sequential()
```

```python
from keras.layers.serialization import activation
#input layer
ann.add(Dense(units=30,activation='relu')) #fully connected layers are called dense layer
```

```python
#first hidden layer
ann.add(Dense(units=8,activation='relu'))
```

```python
#add second hidden layer
ann.add(Dense(units=6,activation='relu'))
```

```python
#third hidden layer
ann.add(Dense(units=4,activation='relu'))
```

```python
#output layer
ann.add(Dense(units=1,activation='sigmoid'))  #In binary classification we use sigmoid in output layer
```

```python
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])  #adaptive momentum=adam(to update weights)
```

```python
ann.fit(X_train,y_train,batch_size=32,epochs=100)
#epoch=trains the same data 100 tyms if we give 100
#batch size= to make training faster we split data into batches and train
```

```python
ann.summary()
```

```python
y_pred=ann.predict(X_test)
y_pred=(y_pred>0.5)
y_pred
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
```
